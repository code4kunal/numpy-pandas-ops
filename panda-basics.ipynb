```python
Table1 = pd.DataFrame({'CustID': ['1001', '1002', '1003', '1004'],
                     'Q1': ['101', '102', '103', '104'],
                     'Q2': ['201', '202', '203', '204']})
   
Table2 = pd.DataFrame({'CustID': ['1001', '1006', '1003', '1004'],
                          'Q3': ['301', '302', '303', '304'],
                          'Q4': ['401', '402', '403', '404']})
```

```python
Table1
```

```python
Table2
```

```python
pd.merge(Table1,Table2,how='outer',on='CustID') #merging two dataframes(merge is wrt to columns)
```

```python
pd.merge(Table1,Table2,how='inner',on='CustID') #exact match to CustID
```

```python
pd.merge(Table1,Table2,how='left',on='CustID') #matches to Table1 wrt to custId
```
